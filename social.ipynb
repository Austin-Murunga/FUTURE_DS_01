```python
#drop redudant column
cols_to_drop = ['Unnamed: 0.1', 'Unnamed: 0']
data = data.drop(columns=cols_to_drop)

#head 
data.head()
```

```python
data.info()
```

```python
#check for platform distribution 
Platform_distr = data['Platform'].value_counts()
Platform_distr
```

```python
#mapping month numbers to uppercase month names
month_mapping = {
    1: 'JAN', 2: 'FEB', 3: 'MAR', 4: 'APR',
    5: 'MAY', 6: 'JUN', 7: 'JUL', 8: 'AUG',
    9: 'SEP', 10: 'OCT', 11: 'NOV', 12: 'DEC'
}

#replace month numbers with their corresponding names
data['Month']  =data['Month'].map(month_mapping)
```

```python
# Normalize the Platform column
data['Platform'] = data['Platform'].str.strip().str.lower()
```

```python
# Clean the 'Country' column to remove extra spaces and unify case
data['Country'] = data['Country'].str.strip().str.upper()
```

Exploring the data

```python
#checking the distribution of entries on the 'Sentiment' column
sentiment_distribution = data['Sentiment'].value_counts()
sentiment_distribution
```

```python
# Calculate sentiment counts
sentiment_counts = data['Sentiment'].value_counts().head(10)

# Plot the counts
plt.figure(figsize=(10, 6))
bars = sentiment_counts.plot(kind='bar', color='skyblue', alpha=0.8)

# Add annotations for counts
for index, value in enumerate(sentiment_counts):
    plt.text(index, value, f'{value}', ha='center', va='bottom', fontsize=10)

plt.title('Top 10 Sentiments by Frequency')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns for summation
if 'Sentiment' in data.columns:
    numeric_data = data.select_dtypes(include=['number'])  # Select numeric columns
    sentiment_sums = data.groupby('Sentiment')[numeric_data.columns].sum()  # Group and sum numeric data

    # Sorting by the total sum of all numeric metrics
    top_10_sentiments = sentiment_sums.sum(axis=1).sort_values(ascending=False).head(10)

    # Plotting the top 10 sentiments
    plt.figure(figsize=(10, 6))
    bars = top_10_sentiments.plot(kind='bar', color='skyblue', alpha=0.8)

    # Adding exact numbers at the top of each bar
    for index, value in enumerate(top_10_sentiments):
        plt.text(index, value, f'{value:.0f}', ha='center', va='bottom', fontsize=10)

    plt.title('Top 10 Sentiments by Total Sum of Numeric Metrics')
    plt.xlabel('Sentiment')
    plt.ylabel('Total Sum of Metrics')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("'Sentiment' column not found in the dataset.")
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    Platform_distr,
    labels=Platform_distr.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue', 'lightgreen', 'salmon'],
    explode=(0.1, 0.1, 0.1),  # Highlight the largest slice
    shadow=True
)
plt.title('Platform Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 Sentiments
top_10_sentiments = data['Sentiment'].value_counts().head(10).index

# Filter data for top 10 sentiments
filtered_data = data[data['Sentiment'].isin(top_10_sentiments)]

# Grouping by Sentiment and Platform
sentiment_platform_dist = filtered_data.groupby(['Sentiment', 'Platform']).size().unstack(fill_value=0)

# Plotting the grouped bar chart
sentiment_platform_dist.plot(
    kind='bar',
    figsize=(12, 8),
    colormap='Set2',
    edgecolor='black'
)
plt.title('Distribution of Top 10 Sentiments Across Platforms')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Platform')
plt.tight_layout()
plt.show()
```
